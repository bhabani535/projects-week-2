```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(dfx)
x
```

```python
x=pd.DataFrame(x,columns=dfx.columns)
x
```

Machine learning

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
def maxr2_score(regr,x,y):
    max_r_score=0
    for r_state in range(42,100):
        x_train, x_test, y_train, y_test = train_test_split(x, y,random_state = r_state,test_size=0.20)
        regr.fit(x_train,y_train)
        y_pred = regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print("r2 score corresponding to ",r_state," is ",r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("max r2 score corresponding to ",final_r_state," is ",max_r_score)
    return final_r_state
```

```python
from sklearn.linear_model import LinearRegression
lnr=LinearRegression()
final_r_state=maxr2_score(lnr,x,y)
```

```python
a_score=cross_val_score(lnr,x,y,cv=10)
print(a_score)
print(a_score.mean())
print(np.std(a_score))
```

```python
#final Model
```

```python
x_train,x_test, y_train, y_test = train_test_split(x, y,random_state = 96,test_size=0.30)
lnr.fit(x_train,y_train)
lnrpred=lnr.predict(x_test)
print('Predicted Deaths: ',lnrpred)
print('Real Deaths: ',y_test)
```

```python
print('R2 score',r2_score(lnrpred,y_test))
print('Mean Absolute Error',mean_absolute_error(lnrpred,y_test))
print('Mean Squared Error',mean_squared_error(lnrpred,y_test))
print('Root Mean Squared Error',np.sqrt(mean_squared_error(lnrpred,y_test)))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=100,random_state=96)
rf.fit(x_train,y_train)
rfpred=rf.predict(x_test)
print('Predicted Deaths: ',rfpred)
print('Real Deaths: ',y_test)
```

```python
print('R2 score',r2_score(rfpred,y_test))
print('Mean Absolute Error',mean_absolute_error(rfpred,y_test))
print('Mean Squared Error',mean_squared_error(rfpred,y_test))
print('Root Mean Squared Error',np.sqrt(mean_squared_error(rfpred,y_test)))
```
